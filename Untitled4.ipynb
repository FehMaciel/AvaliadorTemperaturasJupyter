```python
import csv
import matplotlib.pyplot as plt

with open("Temperaturas.csv", mode='r') as arq:
    leitor = csv.reader(arq, delimiter=';')
    next(leitor)
    temperaturas = [(i[0], int(i[1])) for i in leitor]
    temperatura_media = sum(l[1] for l in temperaturas) / len(temperaturas)
    print(f"{temperatura_media:.2f}")
    temperatura_min = min(temperaturas, key=lambda x: x[1])
    temperatura_max = max(temperaturas, key=lambda x: x[1])
    print(f"Temperatura máxima: {temperatura_max[1]} às {temperatura_max[0]}")
    print(f"Temperatura mínima: {temperatura_min[1]} às {temperatura_min[0]}")
    
    with open("relatorio_temperatura.dat", mode='w') as esq:
        esq.write("======================================\n")
        esq.write("Relatório de Análise de Temperaturas \n")
        esq.write("Aluno: Felippe dos Santos Maciel\n")
        esq.write("Data: 19/04/2023\n")
        esq.write(f"Temperatura média foi {temperatura_media:.2f}\n")
        esq.write(f"Temperatura mínima foi {temperatura_min[1]}\n")
        esq.write(f"Temperatura máxima {temperatura_max[1]}\n")
        esq.write("======================================\n")

        
        plt.figure(figsize=(42, 6))
        plt.plot([t[0] for t in temperaturas], [t[1] for t in temperaturas])
        plt.xlabel("Temperatura")
        plt.ylabel("Hora")
        plt.title("Gráfico de Temperaturas")
        plt.savefig("grafico.png")
        plt.show()
        
    with open("relatorio_temperatura.dat", mode='r') as rela:
        leitor = csv.reader(rela, delimiter=',')
        linhas = 0
        for i in leitor:
            if linhas == 0:
                print(i)
                linhas += 1
            else:
                print(i)
                linhas += 1
```
